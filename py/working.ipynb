```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
train = pd.read_csv('../dat/train.csv')
test = pd.read_csv('../dat/test.csv')
df = pd.concat([train,test])
## Concatenate the two dataframes so that we can engineer features for both of them simultaneously.
## We can later separate them by which ones have NaN id or store columns.
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
## Convert the date column from string to datetime
```

```python
df['year'] = df['date'].map(lambda x:x.year)
df['month'] = df['date'].map(lambda x:x.month)
df['day'] = df['date'].map(lambda x:x.day)
df['dayofweek'] = df['date'].map(lambda x:x.dayofweek)
df['weekofyear'] = df['date'].map(lambda x:x.weekofyear)
```

```python
df.head()
```

```python
## Total sales by day
sales_per_day = df.groupby('date').sum()['sales']
sales_per_day = sales_per_day[sales_per_day > 0]
## Exclude days with 0 sales, since those are all days outside of the data collection window.\
sales_per_day.plot(linewidth = 0.8)
## Looks like a pretty clear seasonal trend, as well as an overall upwards trend.
```

```python
## Total sales by month
sales_per_month = df.groupby('month').sum()['sales']
sales_per_month = sales_per_month[sales_per_month > 0]
sales_per_month.plot.bar()
## Total sales spike in July, lowest around December-February.
```

```python
## Total sales by day of week
sales_per_dayofweek = df.groupby('dayofweek').sum()['sales']
sales_per_dayofweek = sales_per_dayofweek[sales_per_dayofweek > 0]
sales_per_dayofweek.index = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
sales_per_dayofweek.plot.bar()
## Weekends have highest sales.
```

```python
## Daily sales per store
sales_by_store = df.groupby(['date','store']).sum()['sales']
sales_by_store = sales_by_store[sales_by_store > 0].unstack()
sales_by_store.rolling(90).mean().plot.line(figsize = (12,6))
## (90-day rolling average)
## Some stores perform better than others, but it looks like every store follows the same seasonal trend.
```

```python
## Arima model
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.graphics.gofplots import qqplot
from statsmodels.tsa.stattools import adfuller
from pandas.plotting import autocorrelation_plot
import itertools
```

```python
train = df.drop(labels = 'id', axis=1)[pd.isna(df['id'])]
test = df.drop(labels = ['sales','id'], axis=1)[pd.isna(df['sales'])]
```

```python
## Checking Stationarity: Autocorrelation Plot
autocorrelation_plot(pd.Series(data = train.groupby('date').sum()['sales'].values, index = train.groupby('date').sum().index))
```

```python
## Dickey-Fuller test
adf = adfuller(pd.Series(data = train.groupby('date').sum()['sales'].values, index = train.groupby('date').sum().index))
print('ADF Statistic: %f' % adf[0])
print('p-value: %f' % adf[1])
print('Critical Values:')
for key, value in adf[4].items():
    print('\t%s: %.3f' % (key, value))
```

Even though the ADF test indicates stationarity, the graph of the data and the autocorrelation plot make it pretty obvious that there is a seasonality effect and a slight increasing linear trend. Let's see what happens when we apply a lag differencing of 1:

```python
train.groupby('date').sum()['sales'].diff(1).dropna().plot.line(figsize = (12,4))
```

```python
## Dickey-Fuller test
adf = adfuller(pd.Series(data = train.groupby('date').sum()['sales'].diff(1).dropna(), index = train.groupby('date').sum()['sales'].diff(1).dropna().index))
print('ADF Statistic: %f' % adf[0])
print('p-value: %f' % adf[1])
print('Critical Values:')
for key, value in adf[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
train_1 = train.groupby('date').sum()['sales'].diff(1).dropna()
```

We get a very good Dickey-Fuller test. So, we will use the 1-differenced data in an ARIMA model to predict future values.

```python
## Grid search of parameters (p,d,q)
pdq = list(itertools.product(range(0,5),range(0,5),range(0,5)))
min_AIC = float('inf')
min_AIC_params = None
for param in pdq:
    try:
        model = ARIMA(train_1, order = param)
        results = model.fit()
        if results.aic < min_AIC:
            min_AIC = results.aic
            min_AIC_params = param
    except:
        continue
min_AIC_params
## Looks like the optimal parameter is (4,0,4)
```

```python
## Create and fit the model, plot the residuals
model = ARIMA(train_1, order = min_AIC_params)
results = model.fit()
results.resid.plot(figsize = (6,4), linewidth = 0.5)
qqplot(results.resid)
## QQPlot looks normal, residual plot doesn't appear to be biased.
```

```python
pred = results.forecast(90)[0]
```

```python
pred = pd.Series(data = undiff(pred, train.groupby('date').sum()['sales'].tail(1).values[0]), index = pd.date_range(start = pd.to_datetime('2018-01-01'), periods = 90, freq = 'D'))
```

```python
def undiff(series, initial):
    res = list()
    last = initial
    for item in series:
        res.append(initial + item)
        last = initial + item
    return res
```
